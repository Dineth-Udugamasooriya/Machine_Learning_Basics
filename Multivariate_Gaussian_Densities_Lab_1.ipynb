2 Uncertainty in Estimation


```python
MaxTrial = 2000
sampleSizeRange = np.linspace(100, 200, 40) #generating 40 numebrs between 100 and 200 (These will be the different sample sizes used for the analysis.)
print(sampleSizeRange)
plotVar = np.zeros(len(sampleSizeRange)) #to store the calculated variance of variances for each sample size.
for sSize in range(len(sampleSizeRange)):
  numSamples = int(sampleSizeRange[sSize]) #take one value from the array
  # print(numSamples)
  vStrial = np.zeros(MaxTrial)
  for trial in range(MaxTrial): #runs MaxTrial times:
    xx = np.random.randn(numSamples,1) # This generates a random sample of size numSamples from a standard normal distribution (mean=0, variance=1).
    vStrial[trial] = np.var(xx)      #store teh variance of the current sample
  plotVar[sSize] = np.var(vStrial) #After all trials for a specific sample size are done, it calculates the variance of the vStrial array (which contains the variances of each trial) and stores it in the plotVar array for that sample size.

fig, ax = plt.subplots(figsize=(4,4))
ax.set_xlabel('Sample Size')
ax.set_ylabel('Varience of Variences')
ax.set_title('Varience of variences vs. Sample Size')
ax.plot((plotVar))
```

The plot will show how the variance of the estimated variance changes with different sample sizes.
We expect to see a decreasing trend as the sample size increases, indicating that the estimated variance becomes more stable with larger samples.
**This simulation demonstrates the concept that with more data, we can expect our estimates (in this case, variance) to become more accurate and less variable across different trials.**

4. Sampling from a multi-variate Gaussian


```python
C = np.array([[2.0,1.0], [1.0,2]])
A = np.linalg.cholesky(C) #Cholesky decomposition A= [[1.41421356 0.0 ][0.70710678 1.22474487]] , AT = [[1.41421356 0.70710678][0.0         1.22474487]]
print(A @ A.T) #reconstructs C by multiplying A with its transpose A.T (results the initial matrix C)
```

```python
X = np.random.randn(1000,2) #This line generates a random 2D array X of shape (1000, 2). Each row represents a data point with 2 features, and there are 1000 such data points.
Y = X @ A #Since X has shape (1000, 2) and A has shape (2, 2), the matrix multiplication results in a new matrix Y with shape (1000, 2)
print(X.shape)
print(Y.shape)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X[:,0], X[:,1], c="c", s=4) #creates a scatter plot of the data points in X.
ax.scatter(Y[:,0], Y[:,1], c="m", s=4) #creates a scatter plot of the data points in Y.
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
```

5. Distribution of Projections

```python
theta = np.pi/3
u = [np.sin(theta), np.cos(theta)] # u is the unit vector
print("The vector: ", u)
print("Magnitude : ", np.sqrt(u[0]**2 + u[1]**2))
print("Angle of theta in degrees: ", theta*180/np.pi)
```

```python
yp = Y @ u # Y consits 1000 data points. we can get vectors for each data point by multiplying Y with the unit vector eg:- (2 1).([i],[j]) -> 2 rows 1 column = 2i + 1j
print(yp.shape) #1D array
print("Projected variance: ", np.var(yp)) #varience of yp values
```

```python
# Store projected variances in pVars & plot
#
nPoints = 50
pVars = np.zeros(nPoints)
thRange = np.linspace(0, 2*np.pi, nPoints)
for n in range(nPoints):
  theta = thRange[n]
  u = [np.sin(theta), np.cos(theta)]
  pVars[n] = np.var(Y @ u)

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(pVars)
```
